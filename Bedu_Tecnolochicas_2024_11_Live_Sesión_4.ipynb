## Series

```python
# Las Series son secuencias unidimensionales ordenadas de pares índice-valor

# Podemos definir una serie pasándole una lista

serie_1 = pd.Series([1, 2, 3, 4, 5])
serie_1
```

```python
# Hasta arriba tienen el nombre de la Serie
# A la izquierda el índice
# A la derecha los valores
# Hasta abajo el tipo de dato contenido en la Serie
```

### Tipos de datos

- int64: Equivalente a int
- float64: Equivalente a float
- bool: Equivalente a bool (duh)
- object: Equivalente a str, o indica que hay una mezcla de tipos de datos numéricos y no-numéricos en la Serie

```python
# Podemos crear un índice personalizado y ponerle nombre a la Serie

serie_2 = pd.Series([1, 2, 3, 4, 5], index=[100, 102, 103, 104, 105], name='Serie 2')
serie_2
```

```python
# Como las Series también se parecen a los diccionarios, podemos crear una a partir de un diccionario

diccionario = {
    'a': 1,
    'b': 2,
    'c': 3,
    'd': 4,
    'e': 5
  }

serie_3 = pd.Series(diccionario, name="Dict")
serie_3
```

```python
serie_2.loc[105]
```

```python
serie_3.loc['d']
```

## Métodos de indexación

Indexar es "pedirle" a Pandas uno o más valores con el objetivo de filtrar la Serie entera y solamente quedarme con lo que necesito

```python
serie = pd.Series(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'], name="letras")
serie
```

```python
# Podemos pedir varios índices a la vez
# Observen que hay un "doble corchete"
# El primer corchete es el que "loc" necesita para funcionar
# El segundo corchete es una lista que contiene los índices que estoy pidiendo

serie.loc[[0, 1, 2]]
```

```python
serie.loc[[0, 3, 5]]
```

```python
# Los elementos van a ser regresados en el orden que los pidamos

serie.loc[[5, 3, 0]]
```

```python
# También puedo pedir desde el primer elemento hasta un índice que yo defina

serie.loc[:4]
```

```python
# Puedo pedir desde un índice que yo defina hasta el final

serie.loc[6:]
```

```python
# Puedo pedir desde un índice específico hasta otro índice específico

serie.loc[3:7]
```

```python
# Todo funciona igual con índices que no son numéricos

serie_2 = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'], name='Serie 2')
serie_2
```

```python
serie_2.loc[['a', 'd', 'e']]
```

```python
serie_2.loc[['d', 'a']]
```

```python
serie_2.loc[:'c']
```

```python
serie_2.loc['c':]
```

```python
serie_2.loc['b':'d']
```

## DataFrames

```python
import pandas as pd
```

```python
# Una de las maneras que tenemos de definir un DataFrame es a partir de diccionarios
# Cada llave del diccionario es el nombre de una columna
# Cada valor del diccionario es una lista que representa los valores de la columna

datos = {
    'nombre': ['Juan', 'María', 'Pedro', 'Ana', 'Luis'],
    'edad': [25, 30, 35, 40, 45],
    'ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Bilbao']
}
```

```python
# DataFrame lleva mayúscula en la "F"

df = pd.DataFrame(datos)
df
```

```python
# Puedo yo asignarle mi propio índice personalizado

df_2 = pd.DataFrame(datos, index=[100, 101, 102, 103, 104])
df_2
```

```python
# Puedo indexar columnas como si el DataFrame fuera una lista
# Al indexar una columna obtenemos de regreso una Serie

df['nombre']
```

```python
# Podemos también indexar más de una columna

df[['nombre', 'edad']]
```

```python
df[['edad', 'nombre']]
```

```python
# Puedo también indexar por índice, como en las Series

df
```

```python
# También esto me regresa una Serie

df.loc[0]
```

```python
# Puedo pedir más de una fila y me regresa otro DataFrame

df.loc[[0, 4]]
```

```python
# También podemos pedir rangos

df.loc[:2]
```

```python
df.loc[2:]
```

```python
df.loc[1:3]
```

```python
# Puedo pedir un índice y una columna al mismo tiempo, para obtener un solo valor
# Observen que en este caso NO utilizo doble corchete
# Esto se debe a que no le estoy pasando una lista, sino que estoy pidiendo dos diferentes cosas

df.loc[0, 'nombre']
```

```python
# Puedo incluso pedir varios índices y varias columnas
# Observen que para la lista de índices y para la lista de columnas SÍ utilizamos corchetes

df.loc[[0, 4], ['nombre', 'edad']]
```

```python
# Del índice 0, solamente las columnas nombre y edad

df.loc[0, ['nombre', 'edad']]
```

```python
# De los índices 0 y 4, solamente la columna nombre

df.loc[[0, 4], 'nombre']
```

## Manipulación de columnas en un DataFrame

```python
import pandas as pd
```

```python
datos = {
    "Nombre": ["Pepinillo Hernández", "Lúpulo de Dios", "Juan Juon", "Jimmy el Patatas", "Lorenzo Retaguardias"],
    "Cereal favorito": ["Korn Floks", "Verdurinis", "Zumbaritas", "Diabetukis, Papá", "Fibra Máxima 3000"],
    "Hora del desayuno": ["11:00", "07:30", "07:00", "08:30", "09:30"]
}
```

```python
df = pd.DataFrame(datos)
df
```

```python
# Para añadir una nueva columna lo hacemos de forma parecida a con un diccionario
# Asegúrate de que la nueva columna tenga el tamaño adecuado

df["Fruta para acompañar el cereal"] = pd.Series(['pera', 'manzana', 'naranja', 'uva', 'sandía'])
df
```

```python
# Con la misma sintaxis podemos reasignar columnas también

df["Fruta para acompañar el cereal"] = pd.Series(['papaya', 'kiwi', 'plátano', 'fresa', 'coco'])
df
```

```python
# Podemos utilizar el método "drop" para eliminar columnas

df.drop(columns=["Hora del desayuno"])
```

```python
# Estos cambios no son persistentes
# Si observo mi variable "df" verán que aún tengo todas las columnas

df
```

```python
# Para hacer este cambio persistente (es decir, que dure a través del tiempo)
# necesito asignar una copia del resultado a otra variable

df_clean = df.drop(columns=["Hora del desayuno"])
df_clean
```
