## For Loop

```python
# El ciclo for se compone de las siguientes partes:
# Comenzamos con la palabra "for"
# Definimos una variable que va a contener el número de la iteración actual (en este caso le llamamos "i")
# Luego escribimos "in"
# Después definimos cuántas iteraciones vamos a tener. En este caso, range(0, 10) define un rango que comienza en 0 y llega hasta 9, de 1 en 1
# Fíjense que llegamos a 9, aunque escribimos 10
# Después van los dos puntos ":"
# Luego, indentamos para escribir el proceso que se hará "i" número de veces. Todo lo que está dentro del bloque for (indentado) se repetirá
# Al quitar la indentación, definimos el final del for loop

for i in range(0, 10):
  print(i)

print('Fin del for loop')
```

```python
# Puedo, por ejemplo, guardar el número de iteración en una lista

lista_de_numeros = []

for i in range(0, 50):
  lista_de_numeros.append(i)

print(lista_de_numeros)
```

```python
# No necesariamente tengo que usar la variable "i"

numero = 150

for i in range(0, 20):
  numero = numero + 1

print(numero)
```

```python
dict_1 = {
    'llave_1': 'valor_1',
    'llave_2': 'valor_2',
    'llave_3': 'valor_3'
}

# Para iterar por diccionarios, la variable (en este caso "key") contendrá cada una de las llaves de mi diccionario

for key in dict_1:
  print(key)
  print(dict_1[key])
```

```python
lista_1 = ['a', 'b', 'c', 'd', 'e']

# Al iterar por una lista, la variable (en este caso "letra") contendrá cada uno de los elementos de la lista

for letra in lista_1:
  print(letra)
```

## Concatenación de Series

```python
import pandas as pd
```

```python
serie_1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
serie_2 = pd.Series([4, 5, 6], index=['d', 'e', 'f'])
```

```python
serie_1
```

```python
# Puedo concatenar estas series de manera vertical

pd.concat([serie_1, serie_2], axis=0)
```

```python
# También podemos concatenar de manera horizontal

# En este caso sucede algo curioso:
# Como la serie_2 no tiene los índices a, b y c, entonces llena esos datos con NaN
# Como la serie_1 no tiene los índices d, e, f, entonces llena esos datos con NaN

pd.concat([serie_1, serie_2], axis=1)
```

```python
# Podemos nombrar las Series para saber cuál es cuál

pd.concat([serie_1, serie_2], axis=1, keys=['serie_1', 'serie_2'])
```

```python
# Si tenemos dos Series que comparten el mismo índice, esto pasa a la hora de concatenarlas de manera horizontal

serie_3 = pd.Series([7, 8, 9], index=['a', 'b', 'c'])

# En este caso, no hay NaNs porque ambas Series tiene valores válidos para cada uno de los índices

pd.concat([serie_1, serie_3], axis=1)
```

## Concatenación de DataFrames

```python
data_1 = {
    'column_1': [1, 2, 3],
    'column_2': [4, 5, 6]
}

df_1 = pd.DataFrame(data_1, index=['a', 'b', 'c'])
df_1
```

```python
data_2 = {
    'column_1': [7, 8, 9],
    'column_2': [10, 11, 12]
}

df_2 = pd.DataFrame(data_2, index=['d', 'e', 'f'])
df_2
```

```python
# Puedo unirlos verticalmente

# En este caso las columnas se llaman igual, entonces pandas sabe que debe de poner una sobre la otra

pd.concat([df_1, df_2], axis=0)
```

```python
# Qué pasa cuando los concateno horizontalmente

# En este caso no funciona muy bien porque ahora tenemos columnas duplicadas

pd.concat([df_1, df_2], axis=1)
```

```python
# Vamos a probar con un DataFrame que tiene diferentes nombres de columnas, pero el mismo índice que "df_1"

data_3 = {
    'column_3': [7, 8, 9],
    'column_4': [10, 11, 12]
}

df_3 = pd.DataFrame(data_3, index=['a', 'b', 'c'])
df_3
```

```python
pd.concat([df_1, df_3], axis=1)
```

- Cuando tengo columnas distintas, pero indices iguales, funciona muy bien concatenar de manera horizontal
- Cuando tengo índices distintos, pero columnas iguales, funciona muy bien concatenar de manera vertical

NOTA: No quiere decir que no pueda ser útil hacerlo diferente, pero hay que tomar en cuenta que Pandas va a completar los valores faltantes con NaNs. Además, no nos gusta tener columnas con nombres repetidos.

```python
# Por último, podemos concatenar más de 2 Series o DataFrames, simplemente tenemos que añadir todas a la lista

df_total = pd.concat([df_1, df_2, df_3], axis=1)
```

```python
import pandas as pd
import requests
import time
```

```python
api_key = 'Fzr7gDkb1I77NN7qXfOfh4E82UxRhmNDEhVWF3B9'

endpoint = 'https://api.nasa.gov/neo/rest/v1/neo/browse?api_key=' + api_key
payload = {}
```

```python
endpoint
```

```python
dict_datos = {}

for i in range(0, 5):
  try:
    # time.sleep espera por la cantidad de segundos indicada antes de proceder
    # Es una muy buena práctica, cuando estamos haciendo llamadas a una API, esperar un poquito de tiempo entre cada llamada
    # De esta manera evitamos sobrecargar el servidor y meter en líos a la página
    # También evitamos que nos bloqueen
    time.sleep(5)
    # timeout le indica a python que si han pasado "n" segundos y todavía no obtenemos respuesta, lancemos un error
    r = requests.get(endpoint, params=payload, timeout=5)

    if r.status_code == 200:
      json = r.json()
      data = json['near_earth_objects']
      dict_datos[i] = data

      # Recordemos que la llave links tiene los urls que tendría yo que llamar para pedir los siguientes datos; en este caso, el url que quiero es "next"
      new_link = json['links']['next']
      endpoint = new_link
  except:
    # continue lo que hace es pasar a la siguiente iteración de nuestro for loop
    # de esta manera, si algo lanza un error lo único que pasa es que nos saltamos esa llamada y continuamos
    continue
```

```python
lista_df = []

for key in dict_datos:
  normalized = pd.json_normalize(dict_datos[key])
  df = pd.DataFrame(normalized)
  lista_df.append(df)
```

```python
df_completo = pd.concat(lista_df, axis=0).reset_index(drop=True)
```

```python
df_completo
```

```python
df_completo.to_csv("df_completo.csv")
```
