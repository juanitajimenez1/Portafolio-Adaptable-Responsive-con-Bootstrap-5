## Apply

```python
df.head()
```

```python
def years_since_published(value):
  as_datetime = pd.to_datetime(value, unit='ms')
  today = pd.to_datetime('today')
  # Como tengo objetos tipo datetime, puedo literal hacer una resta entre el día de hoy y el día de publicación y obtengo otro objeto al cual
  # le puedo pedir la diferencia obtenida en días
  difference_in_days = (today - as_datetime).days
  # Luego, para convertirla a años, divido entre 365
  return difference_in_days / 365
```

```python
df['published_date.numberLong'].apply(years_since_published)
```

```python
def weeks_on_list_as_percentage_of_maximum(value, max_weeks_on_list):
  as_percentage = value / max_weeks_on_list * 100
  return f'{as_percentage:.2f}%'
```

```python
# Esta estructura:
# args=(df['weeks_on_list.numberInt'].max(),)
# requiere mínimo 2 valores, entonces si solo le voy a pasar uno
# tengo que añadir una coma y luego un espacio vacío

df['weeks_on_list.numberInt'].apply(weeks_on_list_as_percentage_of_maximum, args=(df['weeks_on_list.numberInt'].max(),))
```

```python
def weeks_on_list_as_percentage_of_maximum(value, max_weeks_on_list, final_string):
  as_percentage = value / max_weeks_on_list * 100
  return f'{as_percentage:.2f}% - {final_string}'
```

```python
df['weeks_on_list.numberInt'].apply(weeks_on_list_as_percentage_of_maximum, args=(df['weeks_on_list.numberInt'].max(), "Final"))
```
