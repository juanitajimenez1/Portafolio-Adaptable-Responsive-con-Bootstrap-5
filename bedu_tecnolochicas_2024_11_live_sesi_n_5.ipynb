## NaN o Valores Nulos

```python
datos = {
    'precio': [34, 54, np.nan, np.nan, 56, 12, 34],
    'cantidad_en_stock': [3, 6, 14, np.nan, 5, 2, 10],
    'productos_vendidos': [3, 45, 23, np.nan, 24, 6, np.nan]
}

df = pd.DataFrame(datos, index=["Pokemaster", "Cegatron", "Pikame Mucho", "Lazarillo de Tormes", "Stevie Wonder", "Needle", "El AyMeDuele"])
df
```

```python
# el m√©todo isna() nos regresa False cuando el valor NO es NaN y regresa True cuando el valor S√ç es NaN

df.isna()
```

```python
# True es equivalente a 1
# False es equivalente a 0
# Entonces cuando sumamos estos booleanos, podemos obtener la cantidad de NaNs que tenemos en nuestras columnas o filas

df.isna().sum(axis=0)
```

```python
df.isna().sum(axis=1)
```

```python
# Podemos eliminar columnas o filas que contienen NaNs
# la opci√≥n "any" me elimina una columna o fila que tenga por lo menos 1 NaN

# en dropna, axis=0 quiere decir que va a eliminar filas
df.dropna(axis=0, how="any")
```

```python
# La opci√≥n "all" me elimina una columna o fila en la que TODOS los valores sean NaN
df.dropna(axis=0, how="all")
```

```python
# Al igual que con drop, los cambios solamente se hacen persistentes si los guardamos en una variable

df_2 = df.dropna(axis=0, how="all")
df_2
```

```python
# Puedo crear una columna nueva donde TODOS los valores sean un solo valor de esta manera

df['descuento'] = np.nan
df
```

```python
# Elimina todas las columnas en las que haya por lo menos 1 NaN

df.dropna(axis=1, how="any")
```

```python
df.dropna(axis=1, how="all")
```

```python
# Puedo llenar mis NaNs con un valor que yo elija

df_no_nans = df.dropna(axis=0, how="all")
df_no_nans = df_no_nans.dropna(axis=1, how="all")
df_no_nans
```

```python
# Si asumimos que un NaN en productos_vendidos es porque no se ha vendido nada a√∫n, podr√≠amos rellenar ese valor con un 0

df_no_nans['productos_vendidos'].fillna(0)
```

```python
df_no_nans['productos_vendidos'] = df_no_nans['productos_vendidos'].fillna(0)
df_no_nans
```

```python
# A veces tiene sentido rellenar NaNs con el promedio de los valores de la columna

df_no_nans['precio'].fillna(df['precio'].mean())
```

```python
# Pero tambi√©n a veces es un valor tan importante que si es NaN preferimos deshacernos de toda la fila

df_no_nans.dropna(axis=0, how="any")
```

```python
df
```

```python
# Esto no se est√° aplicando directamente a "df"

df.dropna(axis=1, how="all")
```

```python
# Para "guardar" este cambio, tengo que asignarlo a una nueva variable

df_sin_nans = df.dropna(axis=1, how="all")
df_sin_nans
```

```python
# Puedo usar inplace=True para guardar los cambios en el dataframe original

df.dropna(axis=1, how="all", inplace=True)
df
```

```python
df.dropna(axis=0, how="all", inplace=True)
df
```

```python
df.dropna()
```

```python
df.dropna(subset=['precio'])
```

# Lectura de CSV

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Bedu/Datasets_PDP/melbourne_housing-raw.csv')
df.head()
```

```python
df.shape
```

```python
df.head(5)
```

```python
df.isna().sum()
```

```python
# Asumiendo que la columna YearBuilt no la queremos utilizar, podemos eliminarla

df_2 = df.drop(columns=['YearBuilt']).copy()
df_2.head()
```

```python
df.info()
```

```python
# Asumiendo que necesitamos los valores de BuildingArea, podr√≠amos rellenar los NaNs con la media o la mediana de la columna

df_2['BuildingArea'] = df_2['BuildingArea'].fillna(df['BuildingArea'].median())
df_2.head()
```

```python
df_2.isna().sum()
```

```python
# En el caso de Regionname, podr√≠amos rellenarlo con el valor "Unknown" para no tener que deshacernos de esas filas

df_2['Regionname'] = df_2['Regionname'].fillna('Unknown')
df_2.head()
```

```python
df_2.isna().sum()
```

```python
# Ahora s√≠, del resto eliminemos todas las filas que tienen alg√∫n NaN

df_dropped = df_2.dropna(axis=0, how="any")
df_dropped.head()
```

```python
df_dropped.isna().sum()
```

```python
df_dropped.shape
```

```python
df.shape
```

```python
# Puedo guardar mis datos a un csv

df_dropped.to_csv('melbourne_housing-dropped.csv')
```

```python
df_dropped
```

```python
# Vamos a resetear nuestro √≠ndice para no tener huecos en la numeraci√≥n

df_dropped = df_dropped.reset_index(drop=True)
df_dropped
```

```python
# Puedo tambi√©n cambiar los nombres de mis columnas para que sean m√°s consistentes

df_dropped.columns
```
